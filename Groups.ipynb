```python
import pandas as pd
```

```python
base = pd.read_excel("Copy of Kiso2017Cohort_60month_EvaluationFileFinal_1%_4%_24_08_2022.xlsx", sheet_name="2017cohortBaseOutlierMan")
```

```python
base.head()
```

```python
base = base.iloc[:1224,:]
```

```python
base_sorted = base.sort_values(by='HH Income + Consumption', ascending=False)
```

```python
df_aggregated = base.groupby('Village Name').agg({'HH Income + Consumption': 'mean'}).reset_index()
```

```python
df_aggregated_sorted = df_aggregated.sort_values(by='HH Income + Consumption', ascending=False)
```

```python
df_aggregated_sorted['Group'] = pd.qcut(df_aggregated_sorted['HH Income + Consumption'], 4, labels=['Group 1', 'Group 2', 'Group 3', 'Group 4'])
```

```python
df_aggregated_sorted.to_excel("2017Cohortgroups.xlsx", index=None)
```
