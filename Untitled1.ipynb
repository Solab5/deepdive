```python
mean_performance_per_team_phase = df.groupby(['team_id', 'phase'])['totalingred'].mean().unstack()
mean_performance_per_team_phase
```

```python
mean_performance_per_team_phase['performance_change'] = mean_performance_per_team_phase[2] - mean_performance_per_team_phase[1]
mean_performance_per_team_phase
```

```python
team_group = df.groupby('team_id')['group'].first()
team_group
```

```python
performance_change_with_group = mean_performance_per_team_phase.join(team_group)
performance_change_with_group
```

```python
plt.figure(figsize=(14, 7))
sns.barplot(data=performance_change_with_group, x='group', y='performance_change', ci=None)
plt.title('Mean Performance Change from Phase 1 to Phase 2 by Group')
plt.xlabel('Group')
plt.ylabel('Mean Performance Change')
plt.show()

# Show the mean performance change for each group
performance_change_with_group.groupby('group')['performance_change'].mean()
```

```python
data = df.copy()
```

```python
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd


# Calculate the performance change from Phase 1 to Phase 2 for each team
performance_change = data.groupby(['team_id', 'group', 'phase'])['totalingred'].mean().unstack()
performance_change['change'] = performance_change[2] - performance_change[1]

# Perform ANOVA on the performance changes between phases for the three groups
anova_results = f_oneway(
    performance_change[performance_change.index.get_level_values('group') == 'ai']['change'],
    performance_change[performance_change.index.get_level_values('group') == 'control']['change'],
    performance_change[performance_change.index.get_level_values('group') == 'newhire']['change']
)

# Conduct Tukey's HSD test if ANOVA is significant
if anova_results.pvalue < 0.05:
    # Prepare data for Tukey's HSD
    tukey_data = performance_change.reset_index()
    tukey_results = pairwise_tukeyhsd(
        tukey_data['change'],
        tukey_data['group'],
        alpha=0.05
    )
else:
    tukey_results = "ANOVA result is not significant; no post-hoc test required."

(anova_results, tukey_results if isinstance(tukey_results, str) else tukey_results.summary())
```
